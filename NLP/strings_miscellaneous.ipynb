##### Literal String vs. Regular ExpressionWithout regex=False (default or True): 
- With regex=False: Treats "Merc" as a literal string, not a regular expression (faster and safer if you don't need regex).
- with regex=True: The string "Merc" is treated as a pattern. If it was "Me.c", the dot . acts as a wildcard, potentially matching "Merc", "Me-c", "Meac", etc.
- With regex=False: The string "Merc" is treated as a rigid, literal sequence of characters. The computer looks specifically for 'M' followed by 'e' followed by 'r' followed by 'c'.

```python
df['car_name'].str.replace("Merc", "Mercedes", case=False, regex=False)
```

# Regular Expression re
- Regex = a way to describe patterns in text
- “How would you find all phone numbers in a paragraph?”
- “How would you find emails in a dataset?”

```python
import re
text = "Car prices: 10000, 20000, 30000"
# Find all numbers
numbers = re.findall(r"\d+", text)
numbers_wo_plus = re.findall(r"\d", text) # \d+  → one or more digits. It reads regex like a sentence. It stops at comma(,) beause , is not dight.
print(numbers,'\n')
print(numbers_wo_plus)
```

```python
# extract numbers from car names
import re
cars = ["Mazda RX4", "Duster 360", "Merc 240D"]
for car in cars:
    nums = re.findall(r"\d+", car)
    print(car, "→", nums)
```

```python
# only numbrers as a list
import re

cars = ["Mazda RX4", "Duster 360", "Merc 240D"]
all_numbers = [] # 1. Create an empty list outside the loop

for car in cars:
    nums = re.findall(r"\d+", car)
    all_numbers.append(nums) # 2. Append the found numbers to the list

print(all_numbers)
```

```python
# extract cars with numbers
import re
cars = ["Mazda RX4", "Valiant", "Duster 360"]
filtered = [car for car in cars if re.search(r"\d", car)]
print(filtered)
```

#### Breaking it Down
- for car in cars: This loops through every item in your list (named cars).
- re.search(): This function looks anywhere in the string for a specific pattern.
- r"\d": This is the regex pattern. As you noted earlier, \d stands for single digit (0-9).
- if...: The car is only added to the new list if re.search finds a digit.


```python
# In pandas, str.contains() uses the re (regular expression) engine unless you explicitly tell it not to.
df[df['car_name'].str.contains(r"\d")] 
```

#### Pandas Integration: Methods like .str.contains(), .str.replace(), and .str.extract() are designed to accept regex strings directly.

You only need the explicit import if you are using functions directly from the regex library outside of pandas, such as:
- re.search()
- re.match()
- re.findall()

# NLP
- clean text
- tokenize (split into words)
- count words
- find patterns

```python
texts = [    "Python is easy to learn",     "Data Science is powerful",   "Python makes data analysis easy",  "Machine learning is part of data science"]
texts
```

```python
# Clean
clean_texts = [t.lower() for t in texts]
print(clean_texts)
```

```python
# Tokenization
tokens = [t.split() for t in clean_texts]
print(tokens)
```

```python
# Count words
word_count = sum(len(t.split()) for t in clean_texts)
print("Total words:", word_count)
```

```python
# Most frequent word
from collections import Counter
all_words = " ".join(clean_texts).split()
freq = Counter(all_words)
print(freq.most_common(3),'\n')
print(freq)
```

```python
# sentences with Python
python_sentences = [t for t in clean_texts if "python" in t]
print(python_sentences)
```

```python
# Replace a word
updated = [t.replace("data", "information") for t in clean_texts]
print(updated)
```
