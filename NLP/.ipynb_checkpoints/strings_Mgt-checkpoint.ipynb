# Strings in Python

## Built-in Tools (Standard Library)
### Core str methods : Python’s str class itself has a rich set of methods:
- Case changes: .lower(), .upper(), .title(), .capitalize()
- Whitespace: .strip(), .lstrip(), .rstrip()
- Replace & split: .replace(), .split(), .rsplit(), .join()
- Search: .find(), .rfind(), .startswith(), .endswith()
- Tests: .isalpha(), .isdigit(), .isalnum(), .isspace()

```python
text = "  hello Python world  "

# --- Case Changes ---
print(text.lower())      # "  hello python world  " (all lowercase)
print(text.upper())      # "  HELLO PYTHON WORLD  " (all uppercase)
print(text.title())      # "  Hello Python World  " (caps first letter of every word)
print(text.capitalize()) # "  hello python world  " (caps only the first char of the string, 
                                                    # here first characte is space ' ')

# --- Whitespace Handling ---
print(text.strip())      # "hello Python world" (removes whitespace from both ends)
print(text.lstrip())     # "hello Python world  " (removes leading whitespace)
print(text.rstrip())     # "  hello Python world" (removes trailing whitespace)

# --- Replace & Split ---
csv_data = "apple,banana,cherry"
print(text.replace("Python", "Coding")) # Replaces a substring with another
print(csv_data.split(","))              # ['apple', 'banana', 'cherry'] (turns string into list)
print("-".join(["212", "555", "1212"])) # "212-555-1212" (joins list elements with a separator and converts into 
                                        # a single string)

# --- Search & Inspection ---
filename = "report.pdf"
print(filename.startswith("rep")) # True (checks the beginning)
print(filename.endswith(".pdf"))  # True (checks the ending)
print(text.find("Python"))        # 8 (returns the starting index of the first match)
print(text.rfind("o"))            # 16 (returns starting index of the last match)# scan from the right and finds 
                                  # first'o' and returns index counting normally from the left(16) 

# --- Tests (Returns True/False) ---
print("Python3".isalnum()) # True (all characters are letters or numbers)
print("Python".isalpha())  # True (all characters are letters)
print("12345".isdigit())   # True (all characters are numbers)
print("   ".isspace())     # True (all characters are whitespace)
```

### string module
- Extra constants and helpers:

```python
import string
```

```python
string.ascii_letters   # 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
string.digits          # '0123456789'
```

```python
string.punctuation     # '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
```

## re module (Regular Expressions)
- Powerful pattern matching and text extraction:

```python
import re
```

```python
re.findall(r'\d+', 'Order 123, Item 456')  # ['123', '456']
# Extract only numbers
```

## textwrap
- For wrapping and formatting text:

```python
import textwrap

text = "Python is a high-level, general-purpose programming language."

# 1. It looks for the nearest whitespace or hyphen before the 20th character
# 2. It cuts the string there and inserts a \n (newline)
# 3. It repeats this until the end of the string
wrapped_text = textwrap.fill(text, width=20)

print(wrapped_text)
```

### Utilities for Special Cases

	•	difflib — find differences between sequences/strings.
	•	pyparsing — create grammar rules for parsing text.
	•	unidecode — transliterate Unicode text into ASCII.
