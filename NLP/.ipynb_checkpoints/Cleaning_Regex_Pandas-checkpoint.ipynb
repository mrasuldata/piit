## Python - Regular 
- regex (regular expressions) in Python is super powerful for text cleaning and preprocessing. Let me give you a structured overview with examples.
- https://docs.python.org/3/library/re.html

```python
import re
```

## Basic Functions
- re.findall(pattern, text) → returns all matches
- re.search(pattern, text) → finds the first match
- re.sub(pattern, repl, text) → replaces text
- re.split(pattern, text) → splits text

### Cleaning 

```python
# Remove punctuation
text = "# Hello!!! How are you?? I’m fine..."
cleaned = re.sub(r"[^\w\s]", "", text)
text
```

```python
print(cleaned)   # "Hello How are you I’m fine"
```

```python
#  Lowercasing + remove extra spaces
import re

text = "   THIS    is   a   TEST     "

# EXAMPLE 1: Replacing every single space character
# The \s matches each individual space. Without the '+' sign, 
# every single space in the original text is replaced by one hyphen.
cleaned = re.sub(r"\s", "-", text) 
print(text , '\n', cleaned,'\n******************************************') 

# EXAMPLE 2: Collapsing multiple spaces and chaining methods
# 1. re.sub(r"\s+", "-", text): The '+' means "one or more." 
#    It finds groups of spaces and replaces the whole group with a single '-'.
#    RESULT AT THIS STEP: "-THIS-is-a-TEST-"
# 2. .strip(): Since sub() already replaced the leading/trailing spaces 
#    with hyphens, strip() finds NO spaces to remove. The hyphens remain.
# 3. .lower(): Converts the entire string to lowercase.
cleaned2 = re.sub(r"\s+", "-", text).strip().lower()
print(text , '\n', cleaned2)
```

```python
# Remove numbers
text = "Order #123 was placed on 2025-08-15"

# The \d matches any digit (0-9).
# The + means "one or more" digits in a row (a chunk).
# re.sub finds every chunk of numbers and replaces it with an empty string "".
cleaned = re.sub(r"\d+", "", text)

print(text , '\n', cleaned) 
# Result: "Order # was placed on --"
# Note: The hyphens in the date remain because they are NOT digits (\d).
```

```python
# Extract only emails
text = "Contact us at support@example.com or sales@shop.com"

# BREAKDOWN OF THE EMAIL REGEX:
# 1. [a-zA-Z0-9._%+-]+ -> Matches the "username" part before the @. 
#    It looks for one or more (+) letters, numbers, dots, underscores, etc.
# 2. @                 -> Matches the literal "@" symbol.
# 3. [a-zA-Z0-9.-]+    -> Matches the domain name (like "example" or "shop").
# 4. \.                -> Matches a literal dot "." (the backslash "escapes" it).
# 5. [a-z]{2,}         -> Matches the extension (like "com"). 
#    {2,} means it must be at least 2 characters long (e.g., .com, .org, .edu, .co).

emails = re.findall(r"[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-z]{2,}", text)

print(text, '\n', emails) 
# Result: ['support@example.com', 'sales@shop.com']
```

```python
# Extract only hashtags
import re

text = "Loving #AI and #Python for #DataScience!"

# BREAKDOWN OF THE HASHTAG REGEX:
# 1. #   -> Matches the literal '#' symbol.
# 2. \w  -> Matches any "word character" (letters, numbers, or underscores).
# 3. +   -> The quantifier meaning "one or more." 
#           It tells the regex to keep grabbing characters until it hits 
#           something that isn't a word character (like a space or '!').

hashtags = re.findall(r"#\w+", text)

print(text, '\n', hashtags)
print('*******************************************')
# Result: ['#AI', '#Python', '#DataScience']
# Why it stopped at the !:
# In the word #DataScience!, the regex grabbed #DataScience but stopped right before the !. This is because ! is punctuation, and \w only includes letters, numbers, 
# and underscores. The + quantifier keeps the match going as long as the "word character" rule is met, then ends the match as soon as it hits that exclamation point.
hashtags = re.findall(r"#\w", text) # no '+' afrer \w
print(text, '\n', hashtags)
```

```python
# Remove HTML Tags
import re

text = "<p>Hello <b>World</b></p>"

# BREAKDOWN OF THE HTML TAG REGEX:
# 1. <      -> Matches the literal opening bracket of a tag.
# 2. .      -> Matches ANY character (except a newline).
# 3. *?     -> The "Non-Greedy" quantifier. 
#              '*' means zero or more, but the '?' tells it to stop 
#              at the VERY FIRST '>' it finds.
# 4. >      -> Matches the literal closing bracket of a tag.

cleaned = re.sub(r"<.*?>", "", text)

print(text, '\n', cleaned)   
# Result: "Hello World"
```

```python
# Normalize whitespace (newlines, tabs → space)
import re

text = "Hello\nWorld\tPython"

# re.sub(pattern, replacement, string)
# \s matches any whitespace character (space, newline \n, tab \t, etc.)
# + is a quantifier meaning "one or more"
# This replaces every sequence of whitespace with a single space " "
cleaned = re.sub(r"\s+", " ", text)

print(text)    # Original text with newline and tab
print('***************************************************************')
print(cleaned) # Result: "Hello World Python"
```

##  Regex + Pandas (cleaning columns)

```python
import pandas as pd

df = pd.DataFrame({"text": [
    "Hello!!!",
    "Python   is GREAT???",
    "<p>Data Science</p>",
]})

df
```

```python
df["clean_text"] = df["text"].str.lower()\
                              .str.replace(r"<.*?>", "", regex=True)\
                              .str.replace(r"[^\w\s]", "", regex=True)\
                              .str.replace(r"\s+", " ", regex=True)
df

# This chain of commands cleans the 'text' column and saves it to 'clean_text'
# 1. .str.lower(): Converts all characters to lowercase for uniformity.
# 2. .str.replace(r"<.*?>", "", ...): Removes HTML tags (everything between < and >).
# 3. .str.replace(r"[^\w\s]", "", ...):Removes punctuation. Letters, numbers, and spaces (including tabs and newlines) remain, while symbols like ,, !, ?, # are removed.
#    - [^...] means "not"
#    - \w is alphanumeric, \s is whitespace
#    - So it deletes anything that isn't a letter, number, or space.
# 4. .str.replace(r"\s+", " ", ...): Collapses multiple spaces/tabs/newlines into a single space.
```

# Benefits
- With regex, you can flexibly extract, replace, or normalize text patterns.
- It’s especially useful in NLP, web scraping, and log cleaning.

```python
import re

text = "The price is Rs42000.99"
# Pattern for a dollar amount: a literal $, digits, a literal ., and 2 digits
pattern = r"Rs\d+\.\d{2}"

match = re.search(pattern, text)
if match:
    print(f"Found: {match.group()}") # Output: Found: $42.99
```
