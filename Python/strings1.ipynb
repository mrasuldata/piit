# Regular Expression re
- Regex = a way to describe patterns in text
- “How would you find all phone numbers in a paragraph?”
- “How would you find emails in a dataset?”

# Simple Patterns
## Pattern : Meaning  : Example Match
- \d  : digit  :  5
- \w  : Word (letter\number) : A, 3
- .   :  any character : a, @
- ^   : Starts with : start of the string
- $   : ends with  : end of string

```python
import re
text = "Car prices: 10000, 20000, 30000"
# Find all numbers
numbers = re.findall(r"\d+", text)
print(numbers)
```

-  \d+  → one or more digits
-  read regex like a sentence

```python
# extract numbers from car names
import re
cars = ["Mazda RX4", "Duster 360", "Merc 240D"]
for car in cars:
    nums = re.findall(r"\d+", car)
    print(car, "→", nums)
```

```python
# cars with numbers
import re
cars = ["Mazda RX4", "Valiant", "Duster 360"]
filtered = [car for car in cars if re.search(r"\d", car)]
print(filtered)
```

# NLP
- clean text
- tokenize (split into words)
- count words
- find patterns

```python
texts = [    "Python is easy to learn",     "Data Science is powerful",   "Python makes data analysis easy",  "Machine learning is part of data science"]
texts
```

```python
# Clean
clean_texts = [t.lower() for t in texts]
print(clean_texts)
```

```python
# Tokenisation
tokens = [t.split() for t in clean_texts]
print(tokens)
```

```python
# Count words
word_count = sum(len(t.split()) for t in clean_texts)
print("Total words:", word_count)
```

```python
# Most frequent word
from collections import Counter
all_words = " ".join(clean_texts).split()
freq = Counter(all_words)
print(freq.most_common(3))
```

```python
# sentences with Python
python_sentences = [t for t in clean_texts if "python" in t]
print(python_sentences)
```

```python
# Replace a word
updated = [t.replace("data", "information") for t in clean_texts]
print(updated)
```
