```python
import pandas as pd
import os

# Load the official multi-sheet Excel file
excel_file = r"C:\analytics\data\Superstore\Sample_Superstore.xlsx"

# Use 'calamine' to completely bypass the broken header/footer layouts
xl = pd.ExcelFile(excel_file, engine="calamine")

for sheet_name in xl.sheet_names:
    df = pd.read_excel(excel_file, sheet_name=sheet_name, engine="calamine")
    
    # Save clean data rows without the visual metadata
    base_name = os.path.splitext(excel_file)[0]
    csv_file = f"{base_name}_{sheet_name}.csv"
    
    df.to_csv(csv_file, index=False, encoding="utf-8")
    print(f"Successfully saved: {csv_file}")
```

```python
# Load into DataFrame
orders_df = pd.read_csv(r'C:\analytics\data\Superstore\Sample_Superstore_Orders.csv')
returns_df= pd.read_csv(r"C:\analytics\data\Superstore\Sample_Superstore_Returns.csv")
```

```python
orders_df.info()
```

```python
returns_df.info()
```

```python
# Clean data types and date formats
orders_df['Order Date'] = pd.to_datetime(orders_df['Order Date'])
orders_df['Ship Date'] = pd.to_datetime(orders_df['Ship Date'])
orders_df['Days to Ship'] = (orders_df['Ship Date'] - orders_df['Order Date']).dt.days

# Handle leading zeros for geographic codes
orders_df['Postal Code'] = orders_df['Postal Code'].astype(str).str.zfill(5)

# Simulate Tableau's logical relationship layer
df = pd.merge(orders_df, returns_df, on='Order ID', how='left')
df.to_csv('super_store_tableau_data.csv', index=False)
df['Returned'] = df['Returned'].fillna('No')

print(f"Data Prepared: {len(df)} rows loaded successfully.")
```

#### `Which Top 5 Cities generate the highest total Sales?`

```python
df.groupby('City')['Sales'].sum().nlargest(5)
```

#### `What is the overall distribution of Sales across the 3 Categories?`

```python
df.groupby('Category')['Sales'].sum()
```

#### `What is the Quarter-over-Quarter (QoQ) Sales growth rate?`

```python
qoq = df.resample('QE', on='Order Date')['Sales'].sum().pct_change() * 100
print(qoq)
```

#### `Find the average shipping duration (in days) for each shipping mode across different quarters of the year, and does it spike during Q4?`

```python
print(df.groupby([df['Order Date'].dt.quarter, 'Ship Mode'])['Days to Ship'].mean().unstack())
```

#### `Create a Pareto Analysis showing if 20% of Products generate 80% of Sales.`

```python
prod_sales = df.groupby('Product ID')['Sales'].sum().sort_values(ascending=False).to_frame()
prod_sales['Cum_Pct'] = prod_sales['Sales'].cumsum() / prod_sales['Sales'].sum() * 100
print(prod_sales[prod_sales['Cum_Pct'] <= 80].shape[0] / len(prod_sales) * 100)
```

#### `Find orders that took exactly 0 days to ship but were NOT sent via "Same Day" shipping.`

```python
print(df[(df['Days to Ship'] == 0) & (df['Ship Mode'] != 'Same Day')][['Order ID', 'Ship Mode']])
```

#### `Which states have a negative median profit despite having positive average sales?`

```python
state_stats = df.groupby('State').agg(Avg_Sales=('Sales', 'mean'), Med_Profit=('Profit', 'median'))
print(state_stats[(state_stats['Avg_Sales'] > 0) & (state_stats['Med_Profit'] < 0)])
```

#### `Find "High-Volume Returners" (Customers who returned more than 50% of total ordered items).`

```python
# Create a temporary numeric column for returns
df['Is_Returned'] = (df['Returned'] == 'Yes').astype(int)

# Group and aggregate everything at once
cust_perf = df.groupby('Customer ID').agg(
    Total_Orders=('Order ID', 'count'),
    Total_Returns=('Is_Returned', 'sum')
)

# Calculate rate
cust_perf['Return_Rate'] = cust_perf['Total_Returns'] / cust_perf['Total_Orders']

# Filter
print(cust_perf[cust_perf['Return_Rate'] >0.50].sort_values(by= 'Return_Rate',ascending=False))
```

#### `Find Cross-Category Buyers" who purchased from Technology and Furniture but never Office Supplies.`

```python
# Create a boolean matrix of Customer ID vs Category
cat_matrix = df.groupby(['Customer ID', 'Category']).size().unstack(fill_value=0) > 0

# Vectorized condition check
valid_custs = (
    cat_matrix['Technology'] & 
    cat_matrix['Furniture'] & 
    ~cat_matrix.get('Office Supplies', False)
)

# Get matching customer IDs
cross_buyers = cat_matrix.index[valid_custs]

# Extract unique names
print(df[df['Customer ID'].isin(cross_buyers)]['Customer Name'].unique())
```

#### `What is the average number of days that pass between a customer's 1st purchase and their 2nd purchase?`

```python
# Deduplicate daily orders and ensure chronologically sorted data
cust_orders = df[['Customer ID', 'Order Date']].drop_duplicates().sort_values(['Customer ID', 'Order Date'])

# Extract only the first 2 orders per customer (rank 1 and 2)
first_two = cust_orders.groupby('Customer ID').head(2).copy()

# Calculate days between consecutive dates and take the mean
first_two['Days_Between'] = first_two.groupby('Customer ID')['Order Date'].diff().dt.days
print(first_two['Days_Between'].mean())
```
