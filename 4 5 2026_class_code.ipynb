### Set
- an unordered collection of data that is mutable and does not allow any duplicate element
- Basic uses include membership testing and eliminating duplicate entries
- set ([ x, x, ]) : create like this

```python
groupA = set(['England','USA', 'Spain', 'India'])
groupB = set(['India','France','Poland','India']) #India twice
```

```python
print(type(groupA), type(groupB))
```

```python
print(groupA, "\n", groupB)  #duplicates in groupB removed
```

```python
groupA.union(groupB)
```

```python
groupA.intersection(groupB) #common to both the sets
```

```python
groupA - groupB  #names  in a but not in b
```

```python
groupB - groupA #names  in b but not in a
```

```python
groupA  | groupB #like union
```

```python
groupA  & groupB  # in both, like intersection
```

```python
groupA  ^ groupB  # in A or B but not in Both (india excluded)
```
