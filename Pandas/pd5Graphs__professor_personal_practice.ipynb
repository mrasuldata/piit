## Pandas Graphs

```python
import pandas as pd
from pydataset import data
import matplotlib.pyplot as plt
```

```python
# Set Pandas options for wrapping
pd.set_option('display.max_columns', None)   # Show all columns
pd.set_option('display.width', 100)          # Set line width before wrapping
pd.set_option('display.max_colwidth', None)  # Don't truncate text
```

```python
# Load dataset
air = data('AirPassengers')
air
```

```python
air.info()
```

## Line Plot
- Great for showing time trends

```python
# Line plot of passenger numbers over time
air.plot(y='AirPassengers', x ='time', kind='line', title='Monthly Airline Passenger Numbers')
plt.ylabel("Passengers")
plt.xlabel("Years")
plt.show(); 
```

## Histogram
-  Useful for understanding distribution of a single numeric variable.

```python
mtcars = data('mtcars')
print(mtcars['mpg'].head())
mtcars['mpg'].plot(kind='hist', bins=10, title='MPG Distribution')
plt.xlabel("Miles per gallon")
plt.show()
```

## Barplot
- Group + aggregate + plot = quick summary chart.

```python
mtcars.groupby('gear')['mpg'].mean().plot(kind='bar', title='Average MPG by Gear')
plt.ylabel("Average MPG")
plt.show()
```

The selected code creates a bar chart showing the average miles per gallon (MPG) grouped by the number of gears. Here's a step-by-step explanation:

1. `mtcars.groupby('gear')['mpg'].mean()` - This groups the mtcars DataFrame by the 'gear' column, selects the 'mpg' column from each group, and calculates the mean value for each group.

2. `.plot(kind='bar', title='Average MPG by Gear')` - This creates a bar chart from the grouped data with a title "Average MPG by Gear".

3. `plt.ylabel("Average MPG")` - This adds a label to the y-axis of the plot.

4. `plt.show()` - This displays the plot.

The code effectively visualizes how the average fuel efficiency (MPG) varies across different gear configurations in the mtcars dataset.

## Box Plot
- Compare distributions across categories.

```python
mtcars.boxplot(column='mpg', by='cyl')
plt.title("MPG by Cylinder Count")
plt.suptitle("")  # remove default title
plt.ylabel("Miles per gallon")
plt.show()
```

## Scatter plot
- Spot correlations or relationships between variables.

```python
mtcars.plot(kind='scatter', x='wt', y='mpg', title='Weight vs MPG')
plt.xlabel("Weight (1000 lbs)")
plt.ylabel("Miles per gallon")
plt.show()
```
