```python
url1 = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv&gid={gid1}"
print(url1)
url2 = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv&gid={gid2}"
print(url2)
```

```python
pdata = pd.read_csv(url1)
sdata = pd.read_csv(url2)
```

```python
merged = pd.merge(pdata, sdata, on='rollno' , how = 'inner')
merged.head()
```

```python
subvar = ['maths','english', 'stats','python']
idvar = ['rollno','name','gender','program']
```

```python
merged.groupby("program").mean(numeric_only=True) 
# Calculate mean of rollno(numeric) which is not desireable(numeric_only=True)
```

```python
# Mean marks by program
merged.groupby("program")[subvar].mean(numeric_only=True)
```


This code performs a grouped statistical calculation on a DataFrame. Let me break it down in detail:

1. **`merged`**: This is a pandas DataFrame object that contains the data being analyzed. It likely contains information where each row represents an observation and columns represent different variables.

2. **`.groupby("program")`**: 
   - This method creates a GroupBy object that groups the data in the DataFrame by unique values in the "program" column
   - For example, if "program" contains values like "Science", "Arts", "Engineering", etc., the data will be split into separate groups for each program type
   - The groupby operation doesn't actually compute anything yet; it just defines how the data should be split

3. **`[subvar]`**: 
   - This selects only specific columns from each group
   - `subvar` is likely a list or array of column names (e.g., `["math_score", "reading_score", "writing_score"]`)
   - This filtering step ensures that only the variables of interest are included in the calculation

4. **`.mean(numeric_only=True)`**: 
   - This method calculates the arithmetic mean (average) for each numeric column within each group
   - `numeric_only=True` is a parameter that tells pandas to only include columns with numeric data types in the calculation
   - This prevents errors that would occur if trying to calculate the mean of non-numeric columns (like text)

The final output will be a DataFrame with:
- An index consisting of the unique values from the "program" column
- Columns corresponding to each variable in `subvar`
- Values representing the mean of each variable for each program group

This type of analysis is commonly used to compare average performance or characteristics across different categories or groups.

```python
# Max marks by Gender
merged.groupby("gender")[subvar].max(numeric_only=True)
```

```python
# Multiple aggregations (mean, max, count) with named agg
```

```python
(
    merged
    .groupby("program")
    .agg(
        mean_maths=("maths", "mean"),
        max_maths =("maths", "max"),
        count_rows =("rollno", "count")
    )
) # numeric_only = False by default and not explicitly mentioned as before
```

```python
(
    merged
    .groupby("program")
    .agg(
        mean_maths=("maths", "mean"),
        max_maths =("maths", "max"),
        #count_rows =("rollno", "count")
    )
)  # numeric_only = False by default and not explicitly mentioned as before
```

```python
# Aggregations across all subjects at once
```

```python
(
    merged
    .groupby("program")[subvar]
    .agg(["mean", "median", "max", "min", "count"])
)
```

```python
# Group by Program + Gender
```

```python
(
    merged
    .groupby(["program", "gender"])[subvar]
    .mean(numeric_only=True)
    .round(2)
    .reset_index()
)
```

```python
#  Pivot-style summary (like Excel pivot)
```

```python
pd.pivot_table(
    merged,
    index="program",
    values=subvar,
    aggfunc={"maths": ["mean","max"], "english": ["mean","max"], "stats": "mean", "python": "mean"}
).round(2).reset_index()
```

The selected code creates a pivot table using pandas with the following steps:

1. `pd.pivot_table()` - This function creates a spreadsheet-style pivot table from the DataFrame called `merged`.

2. `index="program"` - Sets the "program" column as the index (row labels) of the pivot table.

3. `values=subvar` - Uses the variable `subvar` to determine which columns to aggregate in the pivot table.

4. `aggfunc={"maths": ["mean","max"], "english": ["mean","max"], "stats": "mean", "python": "mean"}` - Specifies different aggregation functions for different columns:
   - For "maths" column: calculates both mean and maximum values
   - For "english" column: calculates both mean and maximum values
   - For "stats" column: calculates only the mean
   - For "python" column: calculates only the mean

5. `.round(2)` - Rounds all numeric values in the resulting pivot table to 2 decimal places.

6. `.reset_index()` - Converts the index (program column) back to a regular column in the DataFrame.

## Handling missing marks
- By default, Pandas ignores NaN in mean, max, etc. If you want to require all marks present, drop rows first:

```python
merged.dropna(subset=subvar).groupby("program")[subvar].mean()  
# here we removed rows have missing values in any of the subvar columns list
```

```python
merged.groupby("program")[subvar].mean(numeric_only=True)
```

```python
#  Count of students with any missing subject by Program
```

```python
(
    merged
    .assign(AnyMissing = merged[subvar].isna().any(axis=1))
)
```

The selected code is performing operations on a DataFrame called `merged`. Here's a step-by-step explanation:

1. `merged[subvar].isna()` - This checks for missing values (NaN) in the columns specified by the variable `subvar`. The result is a DataFrame of the same shape as `merged[subvar]` with boolean values (True where values are missing, False otherwise).

2. `.any(axis=1)` - This applies the `any()` function along axis=1 (rows), which returns True for each row that has at least one missing value across the specified columns in `subvar`.

3. `.assign(AnyMissing = ...)` - This creates a new column called 'AnyMissing' in the DataFrame, containing the boolean values from the previous step. Each row will have True if any of the columns in `subvar` have missing values for that row, and False otherwise.

4. The entire expression is wrapped in parentheses, suggesting it's part of a larger chain of DataFrame operations.

```python
(
    merged
    .assign(AnyMissing = merged[subvar].isna().any(axis=1))
    .groupby("program")["AnyMissing"]
    .sum()
    .rename("Students_with_Missing")
)
```

The selected code is a chained pandas operation that calculates how many students in each program have missing values. Here's a step-by-step explanation:

1. Starting with a DataFrame called `merged`

2. `.assign(AnyMissing = merged[subvar].isna().any(axis=1))` - Creates a new column called "AnyMissing" that contains True/False values. For each row (student), it checks if there are any missing values (NaN) across the columns specified in the `subvar` variable. If any value is missing, it marks True.

3. `.groupby("program")["AnyMissing"]` - Groups the data by the "program" column and selects the "AnyMissing" column for aggregation.

4. `.sum()` - Counts how many True values exist in each program group (since True is treated as 1 in summation). This gives the count of students with missing values in each program.

5. `.rename("Students_with_Missing")` - Renames the resulting Series to "Students_with_Missing" for clarity.

The final output is a Series showing the count of students with missing values for each program.

## Tricks
- Top Students

```python
def top_scorer(df, col):
    idx = merged[col].idxmax(skipna=True)
    return merged.loc[idx, ["rollno", "name", col]]

tops = {s: top_scorer(merged, s) for s in subvar}
pd.DataFrame(tops)  # dict of small DataFrames
```

The selected code defines a function called `top_scorer` and uses it to create a dictionary of top scorers for different subjects. Here's a step-by-step explanation:

1. `def top_scorer(df, col):` - This defines a function that takes two parameters: a DataFrame `df` and a column name `col`.

2. Inside the function:
   - `idx = merged[col].idxmax(skipna=True)` - Finds the index of the maximum value in the specified column, ignoring NaN values.
   - `return merged.loc[idx, ["rollno", "name", col]]` - Returns a Series containing the roll number, name, and score of the student with the highest score.

3. `tops = {s: top_scorer(merged, s) for s in subvar}` - Creates a dictionary using dictionary comprehension where:
   - Keys are elements from `subvar` (likely subject names)
   - Values are the results of calling `top_scorer` for each subject

4. `pd.DataFrame(tops)` - Converts the dictionary of Series into a DataFrame, where each column represents a different subject's top scorer.

Here's a detailed line-by-line explanation of the selected code:


```python
def top_scorer(df, col):
    idx = merged[col].idxmax(skipna=True)
    return merged.loc[idx, ["rollno", "name", col]]
```

The selected code `return merged.loc[idx, ["rollno", "name", col]]` retrieves specific columns from a DataFrame for a particular row and returns them. Here's what it does step by step:

2. `["rollno", "name", col]` - This list specifies which columns to extract from that row. It selects the "rollno" column, the "name" column, and a third column whose name is stored in the variable `col`.

3. `return` - The function returns this subset of data (a Series or small DataFrame) containing just the three specified columns for the row with the maximum value.

The selected code `pd.DataFrame(tops)` is creating a new pandas DataFrame from a dictionary called `tops`.

When you pass a dictionary to `pd.DataFrame()`:
- If the dictionary contains scalar values, each key becomes a column name and each value becomes a cell value in that column
- If the dictionary contains lists, Series, or arrays, each key becomes a column name and each value becomes the data for that column
- If the dictionary contains other DataFrames (as the comment suggests with "dict of small DataFrames"), pandas will attempt to concatenate them into a single DataFrame
