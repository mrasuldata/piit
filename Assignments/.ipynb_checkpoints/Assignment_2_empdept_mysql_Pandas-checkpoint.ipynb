```python
# function to run the query with parameters or query
def q(sql, **params):
    with engine1.connect() as conn:
        return pd.read_sql(text(sql), conn, params=params)
```

```python
emp = q("SELECT * FROM emp")
emp
```

#### `1. List the names of analysts and salesmen.`

```python
# MYSQL Query
#SELECT ename, job FROM emp WHERE job in('ANALYST','SALESMAN');
# Equivalent pandas code
result = emp[emp['job'].isin(['ANALYST', 'SALESMAN'])][['ename', 'job']]
result
```

#### `2. List details of employees who have joined before 30 Sep 81.`

```python
# MYSQL Query
# SELECT * FROM emp where hiredate < '1981-09-30';
## Equivalent pandas code
emp['hiredate'] = pd.to_datetime(emp['hiredate']) # Convert to datetime first
result = emp.loc[emp['hiredate'] < '1981-09-30']
result
```

#### `3. List names of employees who are not managers.`

```python
# MYSQL Query
#  SELECT ename, job FROM emp WHERE job <> 'MANAGER';
## Equivalent pandas code
emp[emp['job'] != 'MANAGER'][['ename','job']]
```

#### `4. List the names of employees whose employee numbers are 7369, 7521, 7839, 7934, 7788.`

```python
# MYSQL Query
#  SELECT ename FROM emp WHERE empno in('7369', '7521', '7934', '7788');
## Equivalent pandas code
emp[emp['empno'].isin([7369, 7521, 7934, 7788])] [['ename']]
```

#### `5. List employees not belonging to department 30, 40, or 10.`

```python
# MYSQL Query
# SELECT ename FROM emp WHERE deptno NOT IN (30,40,10);
# Equivalent pandas code
emp[~emp['deptno'].isin([30,40,10])][['ename']]
```

#### 6.  `List employee names for those who have joined between 30 June and 31 Dec.'81`

```python
# MYSQL Query
# SELECT ename FROM emp WHERE hiredate between '1981-06-30' and '1981-12-31';
# Equivalent pandas code
start_date = '1981-06-30'
end_date = '1981-12-31'
emp[emp['hiredate'].between(start_date, end_date)][['ename']]
```

#### 7.  `List the different designations in the company.`

```python
# MYSQL Query
# SELECT DISTINCT(job) AS designation FROM emp;
# Equivalent pandas code
emp[['job']].drop_duplicates().rename(columns={'job': 'designation'})                                                                     
```

#### 8.  `List the names of employees who are not eligible for commission.`

```python
# MYSQL Query
# SELECT ename FROM emp WHERE comm IS NULL OR comm = 0;
# Equivalent pandas code
emp[(emp['comm'].isna ()) | ( emp['comm'] == 0) ][['ename']] 
```

#### 9.  `List the name and designation of the employee who does not report to anybody.`

```python
# MYSQL Query
# SELECT ename AS name, job AS designation FROM emp where mgr IS NULL;
# Equivalent pandas code
emp[emp['mgr'].isna()][['ename','job']].rename(columns ={'ename':'name', 'job':'designation'})
```

#### '10. `List the employees not assigned to any department.`

```python
# MYSQL Query
# SELECT ename FROM emp where deptno IS NULL;
# Equivalent pandas code
emp[emp['deptno'].isna()][['ename']]
```

#### 11.  `List the employees who are eligible for commission.`

```python
# MYSQL Query
# SELECT ename FROM emp WHERE comm IS NOT NULL AND comm != 0;
# Equivalent pandas code
emp[emp['comm'].notna() & ( emp['comm'] != 0)][['ename']] 
```

#### 12.  `List employees whose names either start or end with “S”.`

```python
# MYSQL Query
# SELECT ename FROM emp WHERE ename LIKE 'S%' OR ename LIKE '%S';
# Equivalent pandas code
emp[emp['ename'].str.contains(r'^S|S$', case=False, na=False)][['ename']]
```

#### 13.  `List names of employees whose names have “i” as the second character.`

```python
# MYSQL Query
# SELECT ename FROM emp WHERE ename LIKE'_i%';
# Equivalent pandas code
emp[emp['ename'].str.contains(r'^.i', case=False, na=False)][['ename']] 
```

#### 14.  `List the number of employees working with the company.`

```python
# MYSQL Query
# SELECT COUNT(empno) AS Number_of_Employees FROM emp;
# Equivalent pandas code
#emp[['empno']].count().rename({'empno': 'Nunber_of_Employees'})
emp['empno'].agg(Nunber_of_Employees='count')
```

#### 15.  `List the number of designations available in the EMP table.`

```python
# MYSQL Query
# SELECT count(job) AS Number_of_Designation FROM emp;
# Equivalent pandas code
#emp[['job']].count().rename({'job': 'Nunber_of_designations'})
emp['job'].agg(Nunber_of_designations='count')
```

#### 16.  `List the total salaries paid to the employees.`

```python
# MYSQL Query
# SELECT SUM(sal) AS Total_Salaries FROM emp;
# Equivalent pandas code
#Total_Salaries = emp['sal'].sum()
#print(f'Total_Salaries: {Total_Salaries:.2f}')
emp['sal'].agg(Total_Salaries='sum')
```

#### 17.  `List the maximum, minimum and average salary in the company.`

```python
# MYSQL Query
# SELECT MAX(sal) AS Maximum_Salary, MIN(sal) AS Minimum_Salary, AVG(Sal) AS Average_salary FROM emp;
# Equivalent pandas code
emp["sal"].agg(
    Maximum_Salary="max", Minimum_Salary="min", Average_salary="mean"
)
```

#### 18.  `List the maximum salary paid to a salesman`

```python
# MYSQL Query
# SELECT MAX(sal) FROM emp where job = 'salesman';
# Equivalent pandas code
emp[emp['job'] == 'SALESMAN']['sal'].agg(Maximum_Salary='max')
```
