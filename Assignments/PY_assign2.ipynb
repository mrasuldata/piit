```python
import pandas as pd
from pydataset import data

# Load mtcars dataset
df = data('mtcars')
```

```python
df.head()
```

### 4.1 Add Two Columns
- Create a new column mpg_hp_sum = mpg + hp using a lambda function.

```python
df["mpg_hp_sum"] = df.apply(lambda x: x["mpg"] + x["hp"], axis=1)
df.head()
```

### 4.2 Conditional Label
- Create a new column mpg_label:
	•	"High" if mpg > 20
	•	"Low" otherwise
- Use apply() with a lambda.

```python
df["mpg_label"] = df.apply(lambda x: "High" if x["mpg"] > 20 else "Low", axis=1)
df.head()
```

### 4.3 Multiply Column Values
- Double all hp values using map() with a lambda.

```python
df["hp_double"] = df.apply(lambda x: x["mpg"] * 2, axis=1)
df.head()
```

### 4.4 Filter Rows
- Get a list of all mpg values greater than 25 using filter() with a lambda.

```python
mpg_over_25 = list(filter(lambda x: x > 25, df["mpg"]))
mpg_over_25
```

### 4.5 Reduce to Total
- Use reduce() with a lambda to sum all cyl values in the dataset.

```python
from functools import reduce
cyl_sum = reduce(lambda x, y: x + y, df["cyl"])
cyl_sum
```

### 4.6 Sort by Custom Rule
- Sort the dataset by the reverse of the car names’ length using sorted() with a lambda on the index.

```python
# Sort the index labels by length in reverse order
sorted_index = sorted(df.index, key=lambda x: len(str(x)), reverse=True)
print(list(sorted_index))

# Reorder the DataFrame using the sorted index list
df_sorted = df.reindex(sorted_index)
df_sorted.head()
```

### 4.7 Multi-Column Condition
- Create a column efficient_power = "Yes" if mpg > 20 and hp > 100, else "No".|

```python
df["efficient_power"] = df.apply(
    lambda x: "Yes" if x["mpg"] > 20 and x["hp"] > 100 else "No", axis=1)
df.head()
```

### 4.8 Normalize a Column
- Normalize mpg between 0 and 1 using a lambda inside apply():
- {normalized} = (x - min)/(max - min)

```python
mpg_min = df["mpg"].min()
mpg_max = df["mpg"].max()

df["mpg_normalized"] = df["mpg"].apply(
    lambda x: (x - mpg_min) / (mpg_max - mpg_min))
df.head()
```

### 4.9 Extract First Word
- Extract the first word from each car name (index) into a new column car_make using map() with a lambda.

```python
# Extract the first word from the car name index
df['car_make'] = df.index.map(lambda x: str(x).split()[0])

# View the result
print(df[['car_make']].head())
```

### 4.10 Square and Add
- Create a column hp_mpg_calc = (hp² + mpg²) using apply() with a lambda.

```python
df['hp_mpg_calc'] = df.apply(lambda row: row['hp']**2 + row['mpg']**2, axis=1)
df.head()
```

### 4.11 Filter by Col values
- Filter cars with mpg > 25.

```python
# The lambda 'x' represents the entire DataFrame being filtered
high_mpg_cars = df.loc[lambda x: x['mpg'] > 25]
high_mpg_cars
```

### 4.12 Filter and soted carnames by horsepower
- From above return a list of car names sorted by horsepower.

```python
# Extract car names (index) into a list, sorted by horsepower using a lambda key
sorted_cars = sorted(high_mpg_cars.index, key=lambda x: high_mpg_cars.loc[x, 'hp'])
sorted_cars
```
