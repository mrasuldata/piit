```python
import pandas as pd
```

```python
import sqlite3
```

```python
# Connect to a database file (or create it)
conn = sqlite3.connect("piit.db")
```

```python
conn
```

```python
# Create a cursor object
cursor = conn.cursor()
```

```python
cursor.execute("DROP TABLE IF EXISTS students")
```

```python
# Create a table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    fee REAL
);
""")
```

```python
#cursor.execute('show tables') # this line will fail in sqlite3
# The standard SQL way for SQLite
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

tables = cursor.fetchall()
print(tables)
```

```python
# Insert data
cursor.execute("INSERT INTO students (name, fee) VALUES (?, ?)", ("Suroor", 5000))
cursor.execute("INSERT INTO students (name, fee) VALUES (?, ?)", ("Mandal", 6000))
#
```

```python
# Commit changes
conn.commit()
```

```python
# Query data
cursor.execute("SELECT * FROM students")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# Close connection
conn.close()
```
