```python
# Total compensation sal+comm
#empdept['total_comp'] = empdept['sal'] + empdept['comm'] # it is working fine but problematic for NAN values
empdept['total_comp'] = empdept['sal'].add(empdept['comm'], fill_value=0) # Best for skipping NaN implicitly
empdept[['ename','sal','comm','total_comp']].sort_values('total_comp', ascending=False)
# SQL
#SELECT ename, sal, comm, (sal + COALESCE(comm, 0)) AS total_comp FROM empdept ORDER BY total_comp DESC;
```

```python
# Sequencing order number
result = empdept[['ename','sal','comm','total_comp']]\
         .sort_values('total_comp', ascending=False)\
         .reset_index(drop=True)
result.index += 1  # start from 1 instead of 0
result
```

```python
#Pandas Equivalent
result = empdept[['ename', 'sal', 'comm', 'total_comp']].sort_values('total_comp', ascending=False).reset_index(drop=True).rename(lambda x: x + 1)
result
# .rename() method in Pandas targets the index labels (the rows) when you pass a function or a dictionary without specifying an axis

# SQl equivalent
# SELECT ROW_NUMBER() OVER (ORDER BY (e.sal + COALESCE(e.comm, 0)) DESC) AS "index", e.ename, e.sal, e.comm, (e.sal + COALESCE(e.comm, 0)) AS total_comp FROM emp e JOIN 
# dept d ON e.deptno = d.deptno ORDER BY total_comp DESC;
```
