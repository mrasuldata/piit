The command `%load_ext sql` is a Jupyter Notebook "magic" command used to load the ipython-sql extension. Once loaded, it allows you to run SQL queries directly within your notebook cells using the %sql (single-line) or %%sql (multi-line) syntax

```python
# Connect to a database (example: local SQLite file)
%sql sqlite:///my_database.db 
```

%%sql
CREATE TABLE books (id INTEGER PRIMARY KEY, title TEXT, author TEXT);
INSERT INTO books (title, author) VALUES ('1984', 'George Orwell');

```python
%config SqlMagic.style = '_DEPRECATED_DEFAULT' 
```

```python
%%sql
SELECT * FROM books;
```

```python
# show tables
from sqlalchemy import create_engine, inspect

# Create your engine
engine = create_engine("sqlite:///my_database.db")

# Create an inspector
inspector = inspect(engine)

# Get all table names
table_names = inspector.get_table_names()
print(table_names)
```

Use sqlite:/// for a local file(Hard drive) or sqlite:// for an in-memory database.
