```python
df.head()
```

```python
df.dtypes  # dates type NA
```

```python
# dayfirst=False means 1st value is month if the value is between 1-12, otherwise between (11-31) it is day. 01/02/2023 means Jan1, but 25/02/2023 means Feb25.
# errors='coerce' will change the value with NAT(Not A Time)

df['exam_date'] = pd. to_datetime(df['exam_date'], dayfirst=False, errors='coerce') 
df.dtypes 
```
