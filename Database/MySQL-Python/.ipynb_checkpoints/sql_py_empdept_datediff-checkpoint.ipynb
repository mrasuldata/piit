```python
# function to run the query with parameters or query
def q(sql, **params):
    with engine1.connect() as conn:
        return pd.read_sql(text(sql), conn, params=params)
```

```python
import datetime as dt
```

```python
empdept['hiredate'] = pd.to_datetime(empdept['hiredate']) #.dt.date
```

```python
empdept.info()
```

```python
empdept.head()
```

```python
empdept['hireyear'] = empdept['hiredate'].dt.year
empdept.head()
```

```python
empdept[(empdept['hireyear'] >= 1982) & (empdept['dname'] == 'RESEARCH')]
```

```python
empdept[(empdept['hiredate'].dt.year >= 1982) & (empdept['dname'] == 'RESEARCH')]
```

```python
today = pd.to_datetime('today')
today
```

```python
import numpy as np
```

```python
empdept['MonthDiff'] = ((today - empdept['hiredate']) / np.timedelta64(1, 'D')/ 30.44).astype(int) # each month consists of30.44 days
empdept['Month_Served'] = (today.to_period('M') - empdept['hiredate'].dt.to_period('M')).apply(lambda x: x.n)
# The 1-month difference happens because the two formulas measure completely different definitions of time: one calculates physical days elapsed, while the other calculates
# calendar page turns.
empdept['Year_Served'] = (today.to_period('Y') - empdept['hiredate'].dt.to_period('Y')).apply(lambda x: x.n)
empdept['Day_Served'] = (today.to_period('D') - empdept['hiredate'].dt.to_period('D')).apply(lambda x: x.n)
empdept.head()
```
