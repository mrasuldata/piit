```python
# libraries (install if required)
from sqlalchemy import create_engine, text
import pymysql
import pandas as pd
import os
```

```python
# URL-encode special chars in password if any (e.g., # -> %23, @ -> %40)  #Pass@123 - Pass%40123
engine1 = create_engine("mysql+pymysql://root:piit2026@localhost:3306/piit")
# URL-encode special chars in password if any (e.g., # -> %23, @ -> %40)  #Pass@123 - Pass%40123
#engine1 = create_engine("mysql+pymysql://root:piit2026@localhost:3306/piit")
print(engine1)
#setting up connection mysql with user name - piuser, pwd - Pass@123 and to database piit at port no  3306
```

```python
# function to run the query with parameters or query
def q(sql, **params):
    with engine1.connect() as conn:
        return pd.read_sql(text(sql), conn, params=params)
```

```python
q("SELECT * FROM emp")
```

```python
q("select max(sal) from emp")
```

```python
emp = q('select * from emp')
dept = q('select * from dept')
```

```python
empdept= emp.merge(dept, on='deptno', how='inner')
empdept.head()
```

```python
empnull = q('select count(*) from emp where comm is NULL)
```

```python
dept_analysis = empdept.groupby('dname').agg(total_employees=('empno', 'count'), avg_salary=('sal', 'mean')).reset_index()
print(dept_analysis)
```

```python
highest_paid = empdept.sort_values('sal', ascending=False).groupby('dname').first().reset_index()
print(highest_paid[['dname', 'ename', 'sal']])
```

```python
# Calculate mean salary per dept and transform it to map back to original rows
empdept['dept_avg_sal'] = empdept.groupby('dname')['sal'].transform('mean') # avg(sal) by department and create a new column 'dept_avg_sal'

# Filter employees
high_earners = empdept[empdept['sal'] > empdept['dept_avg_sal']]# based on the condition from empdept 
print(high_earners[['ename', 'dname', 'sal', 'dept_avg_sal']])
```
