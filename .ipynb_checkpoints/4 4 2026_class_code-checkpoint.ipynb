## Brackets in Python
- [ ] - square : creating lists, selecting items
- ( ) - round : functions, create tuple object
- { } - curly : set

### Lists
- Lists (list): Ordered, mutable collections of heterogeneous items.

```python
x = [1,2,3,5, 'Dhiraj', 'Alitah', 'Kamrul Hasan', True, False] # heteregenous
```

```python
x
```

```python
print(x)
```

```python
type(x)
```

```python
print(x[2], x[5], x[7]) #ordered, index 
```

```python
x[0]=100  #mutable
print(x)
```

```python
sum([1,7,10])
```

### Dictionary
- Key-value pair combination
- name:"Dhiraj", gender:"M"

```python
dict1 = {'brand':'Honda', 'model':'Jazz', 'year':2017}
dict1
```

```python
print(dict1['model'])
```

### Set
- an unordered collection of data that is mutable and does not allow any duplicate element

```python
groupA = set(['England','USA', 'Spain', 'India'])
groupB = set(['India','France','Poland','India'])
```

```python
print(groupA, "\n", groupB)
```

```python
groupA.union(groupB)
```
