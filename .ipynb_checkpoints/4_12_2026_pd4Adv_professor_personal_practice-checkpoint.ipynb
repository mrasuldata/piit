## When to use each:
- Use `count()` when you need to know how many valid (non-null) values exist
- Use `size()` when you need the total count regardless of null values
- Use `value_counts()` when you need a frequency distribution of values

```python
mtcars.groupby('mpg_level').count()
```

```python
mtcars.groupby('mpg_level')['mpg'].count()
```


The selected code `mtcars.groupby('mpg_level')['mpg'].count()` performs a grouping operation on a DataFrame followed by a count. Here's what it does step by step:

1. `mtcars.groupby('mpg_level')` - Groups the DataFrame `mtcars` by the values in the `mpg_level` column, creating a GroupBy object.

2. `['mpg']` - Selects only the `mpg` column from each group.

3. `.count()` - Counts the number of non-null values in the `mpg` column for each group.

The result will be a Series where:
- The index contains the unique values from the `mpg_level` column
- The values represent the count of non-null `mpg` values in each group

This is commonly used to see how many observations fall into each category or level.

```python
mtcars.groupby('mpg_level').size()
```

The `mpg` column in `mtcars.groupby('mpg_level')['mpg'].count()` is actually not strictly necessary for just counting the number of rows in each group. 

When you use `.count()` on a specific column, it counts the number of non-null values in that column for each group. If you're simply trying to count how many rows are in each group defined by `mpg_level`, any column would work as long as it doesn't have missing values.

You could achieve the same result with:
- `mtcars.groupby('mpg_level').size()` - This directly counts rows without specifying a column
- `mtcars.groupby('mpg_level').count()` - This would return counts for all columns

The specific choice of the `mpg` column might be because:
1. It's related to the grouping variable (`mpg_level`)
2. The author knows it has no missing values
3. It's just a convention or preference of the coder

If all you need is the count of observations in each group, using `.size()` would be more direct and clearer.

```python
mtcars.mpg_level.value_counts()
```

## Comparing `count()`, `value_counts()`, and `size()` in pandas

These three methods serve similar but distinct purposes in pandas:

### count()`
- **Purpose**: Counts non-null values
- **Usage contexts**:
  - On a Series: `series.count()` returns the number of non-null values
  - On a DataFrame: `df.count()` returns count of non-null values for each column
  - After groupby: `df.groupby('col').count()` counts non-null values in each group
- **Key characteristic**: Ignores NaN/None values

### `size()`
- **Purpose**: Counts total number of elements
- **Usage contexts**:
  - On a Series: `series.size` returns total length (including NaN values)
  - On a DataFrame: `df.size` returns total number of elements (rows × columns)
  - After groupby: `df.groupby('col').size()` counts total rows in each group
- **Key characteristic**: Includes NaN/None values in the count

### `value_counts()`
- **Purpose**: Counts occurrences of unique values
- **Usage**: `series.value_counts()` 
- **Returns**: A Series with unique values as index and their frequencies as values
- **Key characteristics**:
  - Only works on Series (not DataFrames)
  - By default sorts by frequency (descending)
  - Has options for normalization, sorting, handling NaN values

### Explanation of `observed=False` vs `observed=True` in Pandas Pivot Tables

The `observed` parameter in `pd.pivot_table()` controls how categorical data is handled, particularly with respect to missing combinations:

### `observed=False` (default)
- Shows **all possible combinations of index/column values**, even if they don't appear in the data.
- Creates a "complete" pivot table with potentially empty cells
- If the index contains categorical data, it will include all category levels defined in the categorical type
- Useful when you want to see all possible combinations and identify missing data

### `observed=True`
- Shows **only combinations that actually appear** in the data
- Creates a more compact pivot table without empty combinations
- Ignores category levels that don't appear in the data
- Useful when you only care about existing data and want a cleaner output

This parameter is particularly important when working with pandas categorical data types, as it determines whether to include all defined category levels or just the ones present in your dataset.

```python
mtcars.pivot_table(values = 'mpg', index='gear', observed=False)
# no aggregation or default aggregation ie mean
# Mean mpg of Gear 3 cars is 16
```

```python
mtcars.pivot_table(values = 'hp', index='cyl', observed=False)
# mean hp of Cyl-4 cars is 82
```

```python
pd.pivot_table(data=mtcars, observed=False,
    index='gear',          # Rows
    values='mpg',          # Values to summarize
    aggfunc='median'         # Aggregation function
)
# Median of Gear-3 cars is 15.5
```

```python
pivot1 = pd.pivot_table(
    data=mtcars, observed= False,
    index='gear',          # Rows
    columns='am',          # Columns
    values='mpg',          # Values to summarize
    aggfunc='mean'         # Aggregation function
)
```

The selected code uses the `melt()` function to reshape the `aq` DataFrame from a wide format to a long format. Here's a step-by-step explanation:

5. `value_name='Value'` - This parameter sets the name of the new column that will contain the values.

```python
pd.melt(frame=aq, id_vars=['Month', 'Day'], 
        value_vars=['Ozone', 'Wind'],
        var_name='Measurement', 
        value_name='value', 
        col_level=None, 
        ignore_index=True) # 306 rows because of two less columns
```

In the pandas.melt() function, the ignore_index parameter determines whether the resulting long-format DataFrame keeps the index from the original wide-format DataFrame. 

- ignore_index=True (Default): The original index is discarded, and the new DataFrame is assigned a standard integer index starting from 0 (e.g., 0, 1, 2, ..., n).
- ignore_index=False: The original index labels are retained in the new DataFrame. Because melting unpivots columns into rows, the original index labels will be repeated for every unpivoted value associated with that original row.

The selected code filters out columns from a DataFrame that have too many missing values. Here's a step-by-step explanation:

1. `threshold = 0.07` - This line sets a threshold value of 7% (not 30% as the comment suggests).

2. `merged.isnull().mean()` - This calculates the proportion of missing values in each column of the DataFrame named `merged`. The `.isnull()` method creates a boolean mask where `True` represents missing values, and `.mean()` calculates the average of these boolean values (treating `True` as 1 and `False` as 0).

3. `merged.loc[:, merged.isnull().mean() <= threshold]` - This uses the `.loc` accessor to select all rows (`:`) and only those columns where the proportion of missing values is less than or equal to the threshold (7%).

4. `merged3C =` - The filtered DataFrame is assigned to a new variable called `merged3C`.

In summary, this code creates a new DataFrame that only includes columns where at least 93% of the values are non-null.

## Dates
- pd.Timestamp(date.today()) → current date
- errors="coerce" → invalid dates become NaT (Not a Time) instead of crashing.
- Pandas auto-detects most formats (YYYY-MM-DD, DD/MM/YYYY, etc.), but you can force a format:
- Subtract DOB from today → 
- To calculate age in years, you need to divide the timedelta by a constant (e.g., 365.25 days) or use dateutil.relativedelta for an exact year/month calculation.

```python
import pandas as pd
from datetime import date
from dateutil.relativedelta import relativedelta
```

```python
merged.head()
```

```python
merged.dtypes
```

```python
merged['dob'] = pd.to_datetime(merged['dob'], errors="coerce")
# pd.to_datetime(merged['dob'], format="%Y-%m-%d")
merged.dtypes
```

```python
merged.head()
```

```python
today = pd.Timestamp(date.today()) # yyyy-mm-dd format
today
```

```python
merged["age"] = (today - merged["dob"]).dt.days // 365    # // as to interpret division
merged.head()
```

The selected code calculates the age in years for each record in the `merged` DataFrame. Here's a step-by-step explanation:

1. `(today - merged["dob"])` - This subtracts the date of birth column ("dob") from the `today` variable (which likely contains the current date), resulting in a timedelta object representing the time difference.

2. `.dt.days` - This accesses the number of days in each timedelta object.

3. `// 365` - The floor division operator (`//`) divides the number of days by 365 and returns the integer result, discarding any remainder. This converts the age from days to approximate years.

4. `merged["age"] = ...` - This assigns the calculated ages to a new column called "age" in the `merged` DataFrame.

5. The comment `# // as to interpret division` explains that floor division is used to get whole years.

```python
merged["ageYM"] = merged["dob"].apply (
    lambda dob: f"{relativedelta(date.today(), dob).years} years, {relativedelta(date.today(), dob).months} months"
    if pd.notnull(dob) else None
)
merged.head()
```

The selected code creates a new column called "ageYM" in the DataFrame "merged" by applying a function to each value in the "dob" (date of birth) column. Here's a step-by-step explanation:

1. `merged["ageYM"] = merged["dob"].apply(...)` - This creates a new column "ageYM" by applying a function to each value in the "dob" column.

2. `lambda dob: ...` - This defines an anonymous function that takes a date of birth value as input.

3. `f"{relativedelta(date.today(), dob).years} years, {relativedelta(date.today(), dob).months} months"` - For each valid date of birth, this calculates the difference between today's date and the birth date using the `relativedelta` function, then formats it as a string showing years and months.

4. `if pd.notnull(dob) else None` - This is a conditional statement that checks if the date of birth is not null. If it's a valid date, it applies the calculation; otherwise, it assigns None to that row.

The code uses `pd.notnull(dob)` instead of something like `merged["dob"].notnull()` or `df["dob"].notnull()` because:

1. Inside the lambda function, `dob` is already a single value from the "dob" column, not the entire column.

2. The lambda function processes one row at a time, so `dob` represents just one date value in each function call.

3. `pd.notnull()` is the appropriate function to check if a single value is not null/NA.

4. Using `merged` or `df` inside the lambda wouldn't make sense because:
   - The lambda only has access to the specific value being processed
   - The lambda doesn't need to reference the entire DataFrame to check if one value is null

This is a common pattern when using `.apply()` with lambda functions - the function operates on individual values, not the whole DataFrame.
