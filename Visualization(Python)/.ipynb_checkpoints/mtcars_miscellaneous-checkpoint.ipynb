## Matplotlib

```python
a = np.linspace(0, 2 * np.pi, 200)
# create 200 point between 0 and 2 x 3.14 = 6.28 
print(a)
print(len(a))
```

```python
# Array x: Using numpy.linspace, the code generates 200 evenly spaced values between 0 and 30 (inclusive). The step size is approximately 0.15075.
# Array y: This contains the sine of each value in array x

import matplotlib.pyplot as plt
x = np.linspace(0, 30, 200)
y = np.sin(x)
print(x[0:5],'\n', y[0:5])
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, c='green', label='Sine Line', linewidth=2)
ax.legend(loc='upper center', bbox_to_anchor=(0.88, 1.1))
plt.tight_layout()
plt.show();
```

#### This line of code moves the legend from its default position inside the box to a specific spot above the plotting area.
Here is what each part does:
- loc='lower center': This defines which part of the legend box is the "handle." By setting it to 'lower center', you are telling Matplotlib to grab the legend by its bottom-middle point.
- bbox_to_anchor=(0.5, 1.0): This tells Matplotlib where to place that "handle" relative to the axes:
- 0.5: 50% across the width (horizontally centered).
- 1.0: 100% of the height (the very top edge of the plot).
- ncol=1: This forces the legend items into a single column. If you had 4 lines and set ncol=4, the legend would stretch out horizontally across the top instead of being a tall vertical list.
- Pro-tip: If the legend looks too "squashed" against the top line, change 1.0 to 1.05 to add a tiny bit of white space.


```python
mtcars = data('mtcars')
mtcars.head(2)
```

```python
gear_counts = mtcars['gear'].value_counts().sort_index()
gear_counts
```

```python
plt.figure(figsize=(10,3))
plt.bar(gear_counts.index.astype(str), gear_counts.values,  color=['g','y','r'], edgecolor="black")
plt.xlabel("Number of Gears")
plt.ylabel("Count of Cars")
plt.title("Gear Counts in mtcars Dataset")
plt.show();
```

```python
# Add labels on top of bars
bars = plt.bar(gear_counts.index.astype(str), gear_counts.values,  color="skyblue", edgecolor="black")
print(bars)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.2, str(height), ha='center', va='bottom', fontsize=8, fontweight='bold', color = 'green')
plt.show();
```

#### `plt.text()` is used to manually place a text string anywhere on your graph using specific coordinates.
- In your code, it’s being used to place a "label" above each bar. Here is the breakdown of the arguments:
Coordinates (x, y):
- bar.get_x() + bar.get_width()/2: This calculates the horizontal center of the bar. It takes the starting left position (get_x) and adds half the width.
height + 0.2: This sets the vertical position. It places the text at the very top of the bar (height) plus a small buffer (0.2) so the text doesn't touch the bar's edge.

##### This calculation finds the exact horizontal center of a bar. Since plt.text centers text on a single point, you need to tell it exactly where the middle of the bar is. 
`bar.get_x()` This gives you the coordinate of the left edge of the bar.
`bar.get_width()` This is the total thickness of the bar.
`/ 2` We only want to move halfway across that thickness.

The Visualization
If a bar starts at x = 2 and is 1.0 wide:
`get_x() = 2
get_width() / 2 = 0.5
Total = 2.5 (The exact middle)`
Without this math, if you just used bar.get_x(), your text labels would be aligned with the left side of the bars instead of sitting neatly in the center.

The Content:
- `str(height): This is the actual text being displayed—in this case, the value of the bar converted to a string.`

Alignment:

- ha='center' (Horizontal Alignment): Ensures the text is centered on your X-coordinate.
- va='bottom' (Vertical Alignment): Sets the "anchor" of the text to its bottom, meaning the text sits on top of the coordinate.

Styling:
- fontsize=10 and fontweight='bold': Standard formatting to make the numbers stand out.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming mtcars is already loaded as a DataFrame
# mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data

gear_counts = mtcars['gear'].value_counts().sort_index()
bars = gear_counts.plot.bar(color="skyblue", edgecolor="black", rot=0)# in pandas
# bars = plt.bar(gear_counts.index.astype(str), gear_counts.values,  color="skyblue", edgecolor="black") # in Matplotlib

plt.xlabel('Gears')
plt.ylabel('Count')
plt.title('Distribution by Number of Gears')

for i, bar in enumerate(bars.patches):
    y_val = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y_val + 0.2, 
             f"{int(y_val)}", ha='center', weight='bold', color ='green', fontsize = 9)

   

plt.show()
```

- Use enumerate(bars) if you used plt.bar() and assigned it to a variable.
- Use enumerate(ax.patches) if you used df.plot.bar() and want to find the rectangles inside the chart area.

## Seaborn
- https://seaborn.pydata.org/tutorial/categorical.html

```python
import seaborn as sns
```

```python
mtcars = data('mtcars')
mtcars.head(2)
```

```python
plt.figure(figsize=(6,5))
ax = sns.countplot(x="gear", hue='gear', data=mtcars, palette="pastel", edgecolor="black", legend=False)
plt.title("Gear Counts in mtcars Dataset")
plt.xlabel("Number of Gears")
plt.ylabel("Count of Cars")
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10, fontweight="bold") 
plt.show();
#hue="gear" assigns the color scheme properly, legend=False prevents a duplicate legend (since x and hue are the same).
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Set a clean style
sns.set_theme(style="whitegrid")

# 2. Plot directly from the raw dataframe
# sns.countplot handles the counting and sorting for you
ax = sns.countplot(data=mtcars, x='gear', hue = 'gear', palette="magma", edgecolor="black", legend=False)

# 3. Use bar_label (No loop needed!)
# This replaces your 'for i, bar in enumerate' logic
for container in ax.containers:
    ax.bar_label(container, weight='bold', color='blue', fontsize=10, padding=3)

# 4. Standard labeling
plt.title('Distribution by Number of Gears', fontweight='bold')
plt.xlabel('Gears')
plt.ylabel('Count')
```

#### Think of `sns.countplot` as a 2-in-1 tool: it does the math (counting) and the drawing (plotting) at the same time.
Here is the breakdown of each part of that line:
- data=mtcars: You are handing Seaborn the whole table (DataFrame). You don't need to pre-slice it or count values yourself.
- x='gear': You are telling Seaborn: "Look at the 'gear' column. Every time you see a unique number (3, 4, or 5), give it a spot on the X-axis and count how many times it appears."
- ax =..... Seaborn returns an "Axes" object (the "painting" inside the frame). We store it in ax so we can use it later to add labels or titles.

##### Why this is different from plt.bar:
In your previous Matplotlib code, you had to manually do value_counts() first.
- Matplotlib: You calculate the numbers 
 You tell the computer to draw those specific numbers.
- Seaborn: You point to a column. The computer counts the numbers and draws them for you. The "Magic" of countplot: It essentially performs mtcars['gear'].value_counts() behind the scenes and maps the result to the bars automatically.
****************************************************************************************************************************************************
- Patches (The Individual "Brick")
A patch is a single shape—in this case, one specific rectangle bar.
What it is: A list of every single rectangle on the screen.
Best for: When you want to treat each bar differently (e.g., "Make the 3rd bar red").
Usage: for bar in ax.patches:
Logic: You have to manually ask each patch for its height or width to do anything with it.
- Containers (The "Palette" or "Group")
A container is a smart wrapper that holds a set of related bars.
What it is: A list that groups bars by their category or series.
Best for: High-level tasks like adding labels.
Usage: for container in ax.containers:
Logic: The container "knows" its data. You don't have to ask for heights; you just tell the container "Label yourself," and it handles the math.
