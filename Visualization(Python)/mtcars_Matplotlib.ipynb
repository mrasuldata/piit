```python
import matplotlib.pyplot as plt
```

```python
# 1. Histogram
plt.figure()
plt.hist(df['mpg'])
plt.title("Distribution of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
# MPG distribution: Mostly clustered between ~18–24 → typical mid-efficiency cars.
```

#### MPG distribution: Mostly clustered between ~18–24 → typical mid-efficiency cars.
#################################################################################################################################


```python
# 2. Scatter
plt.figure()
plt.scatter(df['hp'], df['mpg'])
plt.title("HP vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.show()
```

#### HP vs MPG: Clear negative relationship → higher horsepower → lower fuel efficiency.
############################################################################################

```python
# 3. Boxplot
plt.figure()
df.boxplot(column='mpg', by='cyl')
plt.title("MPG by Cylinders")
plt.suptitle("")
plt.show()
```

#### Cylinders vs MPG: 4-cylinder cars outperform 6 and 8 cylinders in efficiency.
#############################################################################################################3

```python
# 4. Line plot
df_sorted = df.sort_values(by='wt')
plt.figure()
plt.plot(df_sorted['wt'], df_sorted['mpg'])
plt.title("Weight vs MPG")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```

#### Weight vs MPG: Heavier cars tend to have lower MPG → strong inverse trend.
