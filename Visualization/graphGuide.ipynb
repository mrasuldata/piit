- Ask:
    - What do you want to understand from data?
    - Relationship? → scatter plot
    - Distribution? → histogram
    - Comparison? → bar/box plot
- 👉 Students should first choose the plot type, not the library.

# Simple Method 3 Step Rule
- Data → Mapping → Plot
- Example:
    - Data → mpg, hp
    - Mapping → x = hp, y = mpg
    - Plot → scatter

# Learn in Layers
- Layer 1: Pandas (easiest entry)
    - Students focus on data understanding
    - df.plot(x='hp', y='mpg', kind='scatter
    👉 Goal: “quickly visualize data”
- Layer 2: Matplotlib (control) : Now explain:
    - figure
    - axes
    - labels
    - Example
        - plt.scatter(df['hp'], df['mpg'])
        - plt.xlabel("HP")
        - plt.ylabel("MPG")
        - plt.show()
    - 👉 Goal: “customize plots”
- Layer 3: Seaborn (insights) : Introduce:
    - grouping
    - patterns
    - aesthetics
    - Eg
        - sns.scatterplot(data=df, x='hp', y='mpg', hue='cyl')
- Layer 4 : Use ONE dataset throughout (critical)
    - Use mtcars for everything:
        - Histogram → mpg
        - Scatter → hp vs mpg
        - Box → cyl vs mpg
      - 👉 Students connect concepts instead of memorizing syntax
- Layer - 5 :  Teach Interpretation alongside plotting
    - Every plot must answer:
        - What do you see?
        - Why does it happen?
        - What decision can be made?
    - Example:
        - “As horsepower increases, mileage decreases → performance vs efficiency trade-off”
    - 👉 This is where real learning happens

# Students must:
    - choose plots
    - justify choice
    - explain insights
    👉 This builds analytical thinking + plotting skill
- The goal is NOT:
    - “Students know matplotlib/seaborn”
- The goal is:
    - “Students can look at data and decide how to visualize it”
