```python
import pandas as pd
import numpy as np
from pydataset import data
import seaborn as sns
```

```python
df = data('mtcars')
df.head()
```

```python
sns.scatterplot(data=df, x='hp', y='mpg', hue='cyl')
```

```python
sns.regplot(data=df, x='hp', y='mpg')
```

```python
sns.pairplot(df[['mpg','hp','wt','cyl']])
```

- Pandas → fastest for quick checks
- Matplotlib → full control
- Seaborn → best for storytelling + insights
