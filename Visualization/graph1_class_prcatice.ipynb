## Matplotlib

```python
np.linspace(0, 2 * np.pi, 200)
# create 200 point between 0 and 2 x 3.14 = 6.28 
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0, 30, 200)
y = np.sin(x)
print(x[0:5],'\n', y[0:5])
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, c='green', label='Sine Line', linewidth=2)
plt.show();
```

```python
mtcars = data('mtcars')
mtcars.head(2)
```

```python
gear_counts = mtcars['gear'].value_counts().sort_index()
gear_counts
```

```python
plt.figure(figsize=(10,3))
plt.bar(gear_counts.index.astype(str), gear_counts.values,  color=['r','g','y'], edgecolor="black")
plt.xlabel("Number of Gears")
plt.ylabel("Count of Cars")
plt.title("Gear Counts in mtcars Dataset")
plt.show();
```

```python
# Add labels on top of bars
bars = plt.bar(gear_counts.index.astype(str), gear_counts.values,  color="skyblue", edgecolor="black")
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.2, str(height), ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.show();
```

## Seaborn
- https://seaborn.pydata.org/tutorial/categorical.html

```python
import seaborn as sns
```

```python
mtcars = data('mtcars')
mtcars.head(2)
```

```python
plt.figure(figsize=(6,5))
ax = sns.countplot(x="gear", hue='gear', data=mtcars, palette="pastel", edgecolor="black", legend=False)
plt.title("Gear Counts in mtcars Dataset")
plt.xlabel("Number of Gears")
plt.ylabel("Count of Cars")
for p in ax.patches:
    ax.annotate(f"{p.get_height()}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=10, fontweight="bold") 
plt.show();
#hue="gear" assigns the color scheme properly, legend=False prevents a duplicate legend (since x and hue are the same).
```

```python
tips = sns.load_dataset("tips")
sns.catplot(data=tips, x="day", y="total_bill");
```

## Plotly

```python
import plotly.express as px
df = px.data.tips()
fig = px.box(df, y="total_bill")
fig.show()
```

```python
mtcars = data('mtcars')
mtcars.head(2)
```

```python
fig = px.histogram(mtcars, x="gear", color="gear",   title="Gear Counts in mtcars Dataset", 
                   labels={"gear": "Number of Gears"}, text_auto=True)
# Improve layout
fig.update_layout(bargap=0.2, showlegend=False, xaxis_title="Number of Gears", yaxis_title="Count of Cars")
fig.show();
```
